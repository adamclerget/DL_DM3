```python
# Split dataset into training, validation, and test subsets
dataset = ReviewsDataset(use_vocab=True)
train_set, val_set, test_set = random_split(
    dataset, [0.8, 0.1, 0.1], generator=torch.Generator().manual_seed(SEED)
)
```
